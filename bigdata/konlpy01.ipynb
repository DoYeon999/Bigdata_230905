```python
# 한글 처리 작업
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
# 폰트 추출
fontname='c:/Windows/fonts/malgun.ttf'
font_name=font_manager.FontProperties(fname=fontname).get_name()
rc('font',family=font_name)
```

```python
kolaw_tx=Text(okt.nouns(c))
kolaw_tx.plot(30)
plt.show()
```

```python
# 어떤 글자가 몇 번 나타나는지
kv=kolaw_tx.vocab()
kv
```

```python
# 글자 수가 2자 이상인 것만 선택
kv2=dict()

for tag,count in kv.items():
  if len(tag)>1:
    kv2[tag]=count

kv2
```

```python
from wordcloud import WordCloud
font_path='c:/Windows/fonts/malgun.ttf'
wc=WordCloud(width=1000, height=800, background_color='white', font_path=font_path, random_state=0)
plt.imshow(wc.generate_from_frequencies(kv))
# 축 없애기
plt.axis('off')
# 저장하기
plt.savefig("data/wc1.png")
plt.show()
```

```python
from collections import Counter
s2="aaaaabbbbbbccc"
sarr=list(s2)
print(sarr)
c1=Counter(sarr)
print(c1)
```

```python
nouns1=okt.nouns(c)
counter1=Counter(nouns1)
print(counter1)
counter2=counter1.most_common(10)
print(counter2)

counter3=dict()
for tag, count in counter1.items():
  if (len(tag)>1) and (count>1):
    counter3[tag]=count

print(counter3)
```

```python
wc1=WordCloud(width=800, height=600, background_color='white', font_path=font_path)
plt.imshow(wc1.generate_from_frequencies(counter3))
plt.axis('off')
plt.show()
```

# 실습

1. 명사 추출
2. 출혈빈도 계산
3. 단어의 글자수 2자 이상, 출현빈도 2회 이상인 것 선택
4. 출현빈도에 의해 wordCloud 작성

```python
# 1. 명사 추출
okt.nouns(data[:50])
```

```python
# 2. 출현빈도
news_tx=Text(okt.nouns(data))
news_tx.plot(30)
plt.show()
```

```python
ns=news_tx.vocab()
ns
```

```python
# 3. 글자 수가 2자 이상
ns2=dict()

for tag,count in ns.items():
  if len(tag)>1:
    ns2[tag]=count

ns2
```

```python
news=okt.nouns(data)
cnt1=Counter(news)
print(cnt1)
cnt2=cnt1.most_common(10)
print(cnt2)

cnt3=dict()
for tag, count in cnt1.items():
  if (len(tag)>1) and (count>1):
    cnt3[tag]=count

print(cnt3)
```

```python
# 4. 워드클라우드
from wordcloud import WordCloud
font_path='c:/Windows/fonts/malgun.ttf'
wc=WordCloud(width=1000, height=800, background_color='white', font_path=font_path)
plt.imshow(wc.generate_from_frequencies(ns))
plt.axis('off')
plt.savefig("data/test1.png")
plt.show()
```
