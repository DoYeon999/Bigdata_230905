```python
population=pd.read_excel('data/행정구역_시군구_별__성별_인구수_20230907141126.xlsx')
population
```

```python
population['시도'].unique()
```

```python
population['군구'].unique()
```

```python
population['군구']=population['군구'].apply(lambda v:v.strip())
population['군구'].unique()
```

```python
population['시도군구']=population.apply(lambda r:r['시도']+' '+r['군구'], axis=1)
population
```

```python
population=population[population['군구']!='전국']
population=population[population['군구']!='소계']
population
```

```python
population=population.set_index('시도군구')
population
```

```python
addr_polulation=pd.merge(addr_groupby,population, how='inner', 
                         left_index=True, right_index=True)
addr_polulation
```

```python
addr_polulation2=addr_polulation[['시도_x','군구_x','count','총인구수']]
addr_polulation2
```

```python
addr_polulation2=addr_polulation2.rename(columns={'시도_x':'시도','군구_x':'군구','총인구수':'인구수'})
addr_polulation2
```

```python
addr_polulation2['ratio'] = \
addr_polulation2['count'].divide(addr_polulation2['인구수'])*100000
addr_polulation2
```
