```python
import folium
```

```python
m=folium.Map(location=[35.1795543, 129.0756416], zoom_start=18, tiles='Stamen Terrain')
# m.save('data/map1.html')
m
```

location : 지도의 중심좌표(위도, 경도)
zoom_start : 지도의 확대정도(1~18, default:10)
tiles : 지도의 타일 : 'Stamen Terrain'(지형), Open StreetMap(로드맵), 'Stamen Toner'(흑백)

```python
# 좌표 100개 생성
import numpy as np
from folium import plugins
n=100
data=np.array([np.random.uniform(low=35,high=38, size=n),np.random.uniform(low=126,high=129, size=n)]).T
data
# pop은 문자열로 저장되어짐
popup=[str(i) for i in range(1, n+1)]
# 지도 줌레벨에 따라서  합산수 산출 최대시 개별
m=folium.Map(location=[latitude,longitude],zoom_start=7)
# 플러그인 사용해서 데이터좌표 수
plugins.MarkerCluster(data, popups=popup).add_to(m)
m
```

```python
import pandas as pd
df=pd.read_csv('data/CCTV_20190917.csv')
df
```

```python
df1=df.loc[df['카메라대수']>0,['설치목적구분', '소재지도로명주소','위도','경도']]
df1
```

```python
popups=df1['소재지도로명주소'].to_list()
#popups
lat=df1['위도'].to_list()
#lat
lng=df1['경도'].to_list()
#lng
data=np.array([lat, lng]).T
#data
lat_avg=np.array(lat).mean()
lng_avg=np.array(lng).mean()
print(lat_avg, lng_avg)
m=folium.Map(location=[lat_avg, lng_avg],zoom_start=8)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m
```
