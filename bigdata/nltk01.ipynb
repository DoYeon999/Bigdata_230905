# 토큰 생성
- 자연어 문서에서 분석을 위해 긴 문자열을 작은 단위로 나누는 것
- 문장 단위, 단어 단위, 정규표현식으로 나눌 수 있음

```python
# 문장단위 토큰
from nltk.tokenize import sent_tokenize
print(sent_tokenize(emma_raw[:10000])[2])
# 문장
```

```python
# 단어단위 토큰생성
from nltk.tokenize import word_tokenize
print(word_tokenize(emma_raw[10:100]))
# 문제점 : 단어사이에 공백이나, 특수문자
```

```python
#정규표현식 토큰생성
from nltk.tokenize import RegexpTokenizer
retToken=RegexpTokenizer("[\w]+")
# \w -> 문자만 분리
retToken.tokenize(emma_raw[10:100])
# 공백제거 
```

#어간 추출
PorterStemmer,LancasterStemmer 제공
- 어간추출은 단순히 어미만 제거함

```python

from nltk.stem import PorterStemmer,LancasterStemmer
st1=PorterStemmer()
st2=LancasterStemmer()

words=['fly','flies','flying','flew','flown']
print("Porter stemmer:",[st1.stem(w) for w in words])
print("Lancaster Stemmer:",[st2.stem(w) for w in words])
```

# 원형복원
- 같은 의미를 가지는 여러 단어를 사전형으로 통일하는 작업


```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
#  원형복원
```

# 품사 부착
    - 품사는 낱말을 문법적인 기능이나 형태, 뜻에 따라 구분한 것
    - 품사의 예
        - NMP : 단순고유명사
        - VB : 동사
        - VBP : 동사현재형
        - NN : 명사
        - DT : 관형사

```python
from nltk.tag import pos_tag
sentence='Emma refused to permit us to obtai the refuse permit'
tagged_list=pos_tag(word_tokenize(sentence))
tagged_list
# 단어로 다 쪼개서 고유명사 , 동사, 과거형 등등으로 구분 
# pos_tag를 사용해서 쪼개고 난후에 단어들의 품사들을 구별 
# 동사원형으로 만들고 품사 구분
```

```python
nltk.help.upenn_tagset('NNP')
# 각 품사들의 정의 확인 가능
```

```python
noun_list=[t[0] for t in tagged_list if t[1]=='NN'] # else 없으면 for 뒤에 if 붙음
noun_list
# 품사가 명사인 것만 가져와 실제로 값은 t 0만 리스트로 가지고 온다

noun_list2=[]
for t in tagged_list:
  if t[1]=='NN':
    noun_list2.append(t[0])
noun_list2
```

```python
noun_list3= [t[0] if t[1]=='VB' else '-' for t in tagged_list]
# if else문, for문을 대신해서 간단하게 for문을 돌려서 리스트 만들때 간편하게 참고
print(noun_list3)

noun_list4=[]
for t in tagged_list:
  if t[1]=='VB':
    noun_list4.append(t[0])
  else:
    noun_list4.append('-')
print(noun_list4)
```

```python
from nltk.tag import untag
untag_list=untag(tagged_list)
untag_list
# 태깅 해제하고싶다면 untag_list 사용~
# //형태소 분석까지
```

# FreqDist
- 문서에 사용된 단어(토큰)의 사용빈도 정보를 담는 클래스
- vocab() 메소드로 추출

```python
fd=text.vocab()
```

```python
from nltk import FreqDist
stopwords=['Mr.','Mrs.','Miss','Mr','Mrs','Dear','A','No','Ah','Oh']
# 감탄사들도 서치에 걸려서(대문자) 제외
emma_token=pos_tag(retToken.tokenize(emma_raw))
# 나누어진 토큰들에 태그, 형태소가 고유명사이고, 고유명사들 중에서 이런 글자가 아닌것을 뽑을것
name_list=[t[0] for t in emma_token if t[1]=='NNP' and t[0] not in stopwords] 
name_list 
```

```python
fd_name=FreqDist(name_list)
fd_name
```

```python
print(fd_name.N()) # 전체 단어의 수
print(fd_name['Emma']) # 주어진 단어의 빈도 수
print(fd_name.freq('Emma')) # 주어진 단어의 출현빈도의 비율
```

```python
fd_name.most_common(10) # 가장 빈도수가 높은 10개
```
